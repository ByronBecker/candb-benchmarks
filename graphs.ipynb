```python
import matplotlib.pyplot as plt
import csv

plt.figure(figsize=(15,10))
for t in [
    'sib', 'sip', 'si1', 'su1', 'sd1'
]:
    size = []
    totalHeapSize = []

    with open(f'out/{t}.csv') as csvFile:
        stats = csv.reader(csvFile, delimiter=',')
        for i, row in enumerate(stats):
            if i == 0:
                continue
            size.append(int(row[0]))
            totalHeapSize.append(int(row[6]))

    plt.plot(size, totalHeapSize)

plt.xlabel('Size')
plt.ylabel('Total Heap Size')
plt.legend(['Batch', 'Parallel', 'Insert', 'Update', 'Delete'])
plt.show()
```
